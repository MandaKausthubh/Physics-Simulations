# First Order Approximation
Using Euler's method:

$$
    y(x+h) \approx y(x) + y'(x)h = y^*(x+h)
$$

Now for all all values of $y*(x_0+nh) = y^*(x_0+(n-1)h) + kh$. Where $k = y'(x_0 + \Delta) = f(x_0, y^*(x_0))$.
The following code block shows us the approximation working for approximation of the differetial equation:

$$
    y'(x) = -2y(x)
$$



```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math 


# Initializing the list of data points with the initial value:
y_value = [3]
x_value = [0]

y_value1 = [3]
x_value1 = [0]

# Intializing the slope function:
def Y_dash(x_0, y):
    return -2*y
    pass

h = 0.05

while(x_value[-1] < 2):
    y_new = Y_dash(0, y_value[-1])*h + y_value[-1]
    y_value.append(y_new)
    x_value.append(x_value[-1] + h)

h1 = 0.1

while(x_value1[-1] < 2):
    y_new = Y_dash(0, y_value1[-1])*h1 + y_value1[-1]
    y_value1.append(y_new)
    x_value1.append(x_value1[-1] + h1)

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(x_value, y_value, marker='.', linestyle='--', color='b', label='h = 0.05')
plt.plot(x_value1, y_value1, marker='.', linestyle='--', color='r', label='0.1')

# Add labels and title
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Plot of X vs Y')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
