```python
# Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
# Ploting an exponential function

def Differential(x_val, h, y0):
    y_val = np.zeros_like(x_val)
    y_val[0] = y0
    for i in range(1, len(x_val)):
        y_val[i] = -2 * h * y_val[i - 1] + y_val[i-1]
    return y_val 

x_values = np.arange(0,2,0.05)

fig, ax = plt.subplots()
ax.plot(x_values, Differential(x_values, 0.05, 3), '--g')
ax.set_xlabel("x")
ax.set_ylabel("y")
    
```

```python
# Ploting an exponential function

def Differential(x_val, h, y0):
    y_val = np.zeros_like(x_val)
    y_val[0] = y0
    for i in range(1, len(x_val)):
        k = mt.cos(4*x_val[i-1]) - 2*y_val[i-1]
        y_val[i] = y_val[i-1] + h*k
    return y_val 

x_values = np.arange(0,2,0.05)
x_values1 = np.arange(0,2,0.2)
x_values2 = np.arange(0,2,0.8)

fig, ax = plt.subplots()
ax.plot(x_values, Differential(x_values, 0.05, 3), '--g')
ax.plot(x_values1, Differential(x_values1, 0.2, 3), '--b')
ax.plot(x_values2, Differential(x_values2, 0.8, 3), '--r')
ax.set_xlabel("x")
ax.set_ylabel("y")
```
