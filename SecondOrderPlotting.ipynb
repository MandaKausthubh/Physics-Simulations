```python
# What we do here is the same as the previos First Order Approximation but use the slope we get from the half way point to the next prediction

import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
def Differential(x_val, h, y0):
    y_val = np.zeros_like(x_val)
    y_val[0] = y0
    for i in range(1, len(x_val)):
        # Predicting the mid point
        k1 = -2 * y_val[i-1]
        y_mid = y_val[i-1] + k1 * (h/2)
        k2 = -2 * y_mid
        y_new = y_val[i-1] + k2*h
        y_val[i] = y_new
    return y_val


# def Differential1(x_val, h, y0):
#     y_val = np.zeros_like(x_val)
#     y_val[0] = y0
#     for i in range(1, len(x_val)):
#         k = mt.cos(4*x_val[i-1]) - 2*y_val[i-1]
#         y_val[i] = y_val[i-1] + h*k
#     return y_val 


def Differential1(x_val, h, y0):
    y_val = np.zeros_like(x_val)
    y_val[0] = y0
    for i in range(1, len(x_val)):
        y_val[i] = -2 * h * y_val[i - 1] + y_val[i-1]
    return y_val 

x_values = np.arange(0, 2, 0.2)

fig, ax = plt.subplots()
ax.plot(x_values, Differential(x_values, 0.2, 3), '--b', label=r"2^{nd} order")
ax.plot(x_values, Differential1(x_values, 0.2, 3), '.-r', label = r"1^st order")

ax.legend()
```
